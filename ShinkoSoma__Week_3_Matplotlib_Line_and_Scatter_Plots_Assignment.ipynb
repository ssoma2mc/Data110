### 2. Basic Line Plot
- Create a plot of the quadratic function $f(x) = x^2 - 2x + 1$ over the range $[-2, 4]$
- Use `plt.plot()` to create the plot

````python
import numpy as np

# Generate values for x in the range [-2, 4]
x = np.linspace(-2, 4, 400)

# Compute the corresponding y values for the quadratic function f(x) = x^2 - 2x + 1
y = x**2 - 2*x + 1

# Create the plot
plt.plot(x, y, label="f(x) = x^2 - 2x + 1", color='b')

# Add labels and a title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^2 - 2x + 1')

# Show the legend
plt.legend()

# Display the plot
plt.show()
````

### 3. Multiple Lines
- Create a plot with the following three functions on the same graph:
  - $f(x) = x^3$
  - $g(x) = e^x$
  - $h(x) = \log(x)$ (for $x > 0$)
- Use different colors for each line
- Add a legend to distinguish between the functions

````python
# Generate values for x
x1 = np.linspace(-3, 3, 800)
x2 = np.linspace(0.01,3, 400)  # Starting from 0.01 to avoid log(0)
# Compute the corresponding y values for each function
f_x = x1**3
g_x = np.exp(x1)
h_x = np.log(x2)

# Create the plot for all three functions
plt.plot(x1, f_x, label="f(x) = x^3", color='r')  # Red line for x^3
plt.plot(x1, g_x, label="g(x) = e^x", color='g')  # Green line for e^x
plt.plot(x2, h_x, label="h(x) = log(x)", color='b')  # Blue line for log(x)

# Add the 'o' marker at x = 0 for h(x) with white fill and blue edge
plt.scatter(0, np.log(0.01), color='white', edgecolors='blue', s=50, zorder=5, linewidths=1)

# Add labels and a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiple Functions: $f(x) = x^3$, $g(x) = e^x$, $h(x) = log(x)$')

plt.xlim([-3, 3])
plt.ylim([-15, 15])

# Show the legend
plt.legend()

# Display the plot
plt.show()
````

### 5. Axes Limits and Scaling
- Plot the function $f(x) = \frac{1}{x^2 + 1}$ over the range $[-10, 10]$
- Set appropriate custom x and y limits to showcase the function's behavior
- Create another plot of the same function with reversed x and y axes
- Demonstrate the use of `plt.axis('tight')` and `plt.axis('equal')`

````python
# Generate values for x in the range [-10, 10]
x = np.linspace(-10, 10, 400)

# Compute the corresponding y values for f(x) = 1 / (x^2 + 1)
y = 1 / (x**2 + 1)

# 1. Plot with custom x and y limits
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$f(x) = \\frac{1}{x^2 + 1}$")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $f(x) = \\frac{1}{x^2 + 1}$')

# Set custom limits for x and y
plt.xlim(-10, 10)
plt.ylim(0, 1)

# Add a legend
plt.legend()

# Display the plot
plt.show()

# 2. Reversed Axes Plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$f(x) = \\frac{1}{x^2 + 1}$")
plt.xlabel('y')
plt.ylabel('f(y)')
plt.title('Reversed Axes Plot')

# Reverse the axes
plt.gca().invert_xaxis()  # Reverse x-axis
plt.gca().invert_yaxis()  # Reverse y-axis

# Add a legend
plt.legend()

# Display the plot
plt.show()

# 3. Demonstrating `plt.axis('tight')`
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$f(x) = \\frac{1}{x^2 + 1}$")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot with `axis(\'tight\')`')

# Use `plt.axis('tight')` to automatically adjust the axes to fit the data
plt.axis('tight')

# Add a legend
plt.legend()

# Display the plot
plt.show()

# 4. Demonstrating `plt.axis('equal')`
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$f(x) = \\frac{1}{x^2 + 1}$")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot with `axis(\'equal\')`')

# Use `plt.axis('equal')` to make the scales of both axes equal
plt.axis('equal')

# Add a legend
plt.legend()

# Display the plot
plt.show()
````

````python
plt.style.use('seaborn-v0_8-darkgrid')

# If I don't use h(x)=sin(x)/x, the code is not difficult.
# Python ignore x=0 and generate the graph of h(x) = sin (x)/x, but this is not right.
# To handle ùë•=0, appropriately, I use 'sinc' function for h(x).

# Define x values (in the range from -10 to 10, excluding the region near zero for tan(x) to avoid discontinuity)
x = np.linspace(-10, 10, 400)
x_sinc = np.linspace(-10, 10, 400)

# Avoid division by zero in h(x) = sin(x)/x at x = 0 by manually setting the value at x=0 to 1
y_sinc = np.sin(x_sinc) / x_sinc
y_sinc[x_sinc == 0] = 1  # Handling the singularity at x=0

# Define the functions
f_x = np.tan(x)  # tan(x)
g_x = x * np.exp(-x**2)  # x * e^(-x^2)
h_x = np.sin(x_sinc) / x_sinc  # sin(x)/x (sinc function)
j_x = 1 / (1 + np.exp(-x))  # logistic function
i_x = np.arctan(x)  # arctan(x)

# Scatter points for selected x values
x_scatter = np.arange(-10, 10, 1)
y_scatter_f = np.tan(x_scatter)
y_scatter_g = x_scatter * np.exp(-x_scatter**2)
y_scatter_h = np.sin(x_scatter) / x_scatter
y_scatter_j = 1 / (1 + np.exp(-x_scatter))
y_scatter_i = np.arctan(x_scatter)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the functions with different colors and styles
plt.plot(x, f_x, label='$f(x) = \\tan(x)$', color='blue', linestyle='-', linewidth=2)
plt.plot(x, g_x, label='$g(x) = x e^{-x^2}$', color='red', linestyle='--', linewidth=2)
plt.plot(x_sinc, h_x, label='$h(x) = \\frac{\\sin(x)}{x}$', color='green', linestyle='-.', linewidth=2)
plt.plot(x, j_x, label='$j(x) = \\frac{1}{1 + e^{-x}}$', color='purple', linestyle=':', linewidth=2)
plt.plot(x, i_x, label='$i(x) = \\arctan(x)$', color='orange', linestyle='-', linewidth=2)

# Scatter plot points for selected x values
plt.scatter(x_scatter, y_scatter_f, color='blue', marker='o', s=50, label='Scatter f(x)', zorder=5)
plt.scatter(x_scatter, y_scatter_g, color='red', marker='s', s=50, label='Scatter g(x)', zorder=5)
plt.scatter(x_scatter, y_scatter_h, color='green', marker='^', s=50, label='Scatter h(x)', zorder=5)
plt.scatter(x_scatter, y_scatter_j, color='purple', marker='D', s=50, label='Scatter j(x)', zorder=5)
plt.scatter(x_scatter, y_scatter_i, color='orange', marker='x', s=50, label='Scatter i(x)', zorder=5)

# Customize plot limits
plt.xlim([-10, 10])
plt.ylim([-2, 2])

# Add labels and title
plt.title('Custom Plot with Multiple Functions', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)

# Add legend outside the plot area
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
````

````python
# I discuss about h_x = np.sin(x) / x vs j_x = 1 / (1 + np.exp(-x)).

import seaborn as sns

# Widened the x-range to show the behavior (convergence) of the functions.
x = np.linspace(-30, 30, 400)
x_sinc = np.linspace(-30, 30, 400)

# Avoid division by zero in h(x) = sin(x)/x at x = 0 by manually setting the value at x=0 to 1
y_sinc = np.sin(x_sinc) / x_sinc
y_sinc[x_sinc == 0] = 1  # Handling the singularity at x=0

# Define the functions
g_x = x * np.exp(-x**2)  # x * e^(-x^2)
h_x = np.sin(x_sinc) / x_sinc  # sin(x)/x (sinc function)

# Scatter points for selected x values
x_scatter = np.arange(-30, 30, 1)
y_scatter_g = x_scatter * np.exp(-x_scatter**2)
y_scatter_h = np.sin(x_scatter) / x_scatter

# Create the plot
plt.figure(figsize=(10, 6))

# Used a darker palette to match the seaborn-darkgrid style. generate 8 colors in dark palette.
palette = sns.color_palette("dark6", 8)

# Made the lines simpler and more professional. Used the same colors for the smaller markers, and added transparency to the lines so they remain visible.
plt.plot(x, g_x, label='$g(x) = x e^{-x^2}$', color=palette[0], linestyle='-', linewidth=2, alpha=0.9)
plt.plot(x_sinc, h_x, label='$h(x) = \\frac{\\sin(x)}{x}$', color=palette[1], linestyle='--', linewidth=2, alpha=0.9)

# Chose smaller markers to keep the plot professional and simple. Decided that smaller markers are better to show the convergence of the functions.
plt.scatter(x_scatter, y_scatter_g, color=palette[0], marker='o', s=20, label='Scatter g(x)', zorder=5)
plt.scatter(x_scatter, y_scatter_h, color=palette[1], marker='x', s=20, label='Scatter h(x)', zorder=5)

# Customize plot limits
plt.xlim([-30, 30])
plt.ylim([-2, 2])

# Add labels and title
plt.title('$g(x) = x e^{-x^2}$ vs $h(x) = \\frac{\\sin(x)}{x}$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)

# Add legend outside the plot area
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
````

### 1. Functions Chosen:
- **$g(x) = x e^{-x^2}$**: This function describes a Gaussian-shaped curve that decreases exponentially as x moves away from zero in both directions. The function approaches zero as x approaches to infty, or  x approaches to - infty.
- **$h(x) = \frac{\sin(x)}{x}$**: The logistic function is a sigmoidal curve that asymptotically approaches 1 as x approaches to infty, and approaches 0 as x approaches to - infty. Unlike the exponential decay in g(x), the convergence here is more gradual, resembling a smooth S-shaped curve.

### 2. Graph Adjustments:
- The x-range has been widened to show the behavior (convergence) of the functions more clearly. This allows us to observe the overall trends in both g(x) and h(x) as x approaches larger values.
- The y-limits are properly adjusted to ensure that the full range of function values is visible within the plot.
- A color palette from Seaborn "dark6" was used to differentiate the functions visually. This helps to create clear distinctions between the two functions.

**Reflection on Graph in Task 9**

The graph in Task 9 follows a common structure frequently encountered in fields like medicine and economics, presenting actual data trends. In this case, I have assumed the graph represents the progression of COVID-19 infections.

Graphs utilized in academic research and publications, such as those for journal submissions, are often created in large formats like A0. This raises the question of which style‚Äî**default**, **seaborn-whitegrid**, or **seaborn-darkgrid**‚Äîwould be most suitable when preparing for a conference presentation.

1. **Default Style**  
The default style (A1) offers a clean and minimalistic design, ideal for general-purpose plots where the focus remains on the data itself, free from distractions. This style is excellent for publication-ready graphics. However, for a conference presentation, it has its limitations, as the absence of gridlines can make it harder for the audience to follow the data trends, and the borders can become somewhat distracting, adding unnecessary visual noise.

2. **Seaborn-Whitegrid Style**  
In contrast, the **seaborn-whitegrid** style (A2) features a clear white background with subtle gridlines, which enhance the readability of the graph without overpowering the data. The gridlines help viewers track the values with ease, making the key trends and peak points stand out more prominently. This style provides an elegant and clear presentation, especially suitable for data-focused conference talks.

3. **Seaborn-Darkgrid Style**  
The **seaborn-darkgrid** style (A3) uses a soft purple-gray background, which was visually appealing when working on a light computer display. However, when I attempted to simulate the environment of a conference presentation (B3), particularly in dark-lit rooms, the dark background interfered with the graph's readability. The black border, though intended to complement the dark theme, only exacerbated this issue, making the background color more distracting rather than enhancing the overall clarity of the graph.

**Conclusion**  
In conclusion, for conference presentations, especially in dark conference halls, the **seaborn-whitegrid** style proves to be the most effective. It strikes a balance between clarity and aesthetic appeal, ensuring the graph remains easily readable and visually impactful without any unnecessary distractions.
