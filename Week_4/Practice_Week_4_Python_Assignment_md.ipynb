```python
# Data for the crimes (Offenses per 100,000 People 2023)
crimes = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
offenses_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
change_in_offenses = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

# Sort the data by change_in_offenses in descending order
sorted_crimes, sorted_offenses_2023, sorted_change_in_offenses = zip(
    *sorted(zip(crimes, offenses_2023, change_in_offenses), key=lambda x: x[2], reverse=True)
)

# Colors in the specified order
orange_1 = "#C77C4A"  # Dark Orange
orange_2 = "#E4A96B"  # Light Orange

blue_1 = "#1F3A52"  # Dark Blue
blue_2 = "#2C5A83"  # Darker Blue
blue_3 = "#436D9F" # Blue
blue_4 = "#5A8BCA"  # Lighter Blue
blue_5 = "#6CB4E1"  # Light Blue

# Assign colors based on the given order (with specific mapping)
#print(sorted_crimes) Mannuary adjusted
#('Motor Vehicle Theft', 'Robbery', 'Aggravated Assault', 'Larceny', 'Burglary', 'Rape (revised definition)', 'Murder')
colors = [blue_5, blue_4, orange_2, blue_3, blue_2, blue_1, orange_1]

# Create the horizontal bar plot using matplotlib
plt.figure(figsize=(10, 6))

# Create the bars with the selected colors
bars = plt.barh(sorted_crimes, sorted_change_in_offenses, color=colors)

# Add annotations to each bar for clarity
for index, value in enumerate(sorted_change_in_offenses):
    plt.text(value + 0.5, index, f'{value:.1f}', va='center', color='black', fontweight='bold')

# Customize the plot
plt.xlabel('Change in Offenses per 100,000 People (2022-2023)', fontsize=12, fontweight='bold')
plt.ylabel('Type of Crime', fontsize=12, fontweight='bold')
plt.title('Change in Crime Rates in 2023 (Offenses per 100,000 People)', fontsize=14, fontweight='bold')
plt.grid(True, axis='x', linestyle='--', linewidth=0.7)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()




```

```python
# I have made the graph for additional task 1.
# I rewrite the graph using 'seaborn-v0_8-white grid' and 'Seaborn color palettes'

plt.style.use('seaborn-v0_8-whitegrid')

current_palette = sns.color_palette("dark", n_colors=10)
#sns.palplot(current_palette)

# Data for the crimes (Offenses per 100,000 People 2023)
crimes = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
offenses_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
change_in_offenses = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

# Calculate the 2022 offenses based on the change
offenses_2022 = [offense / (1 + (change / 100)) for offense, change in zip(offenses_2023, change_in_offenses)]

# Sort the data by offenses_2023 in ascending order
sorted_crimes, sorted_offenses_2022, sorted_offenses_2023 = zip(
    *sorted(zip(crimes, offenses_2022, offenses_2023), key=lambda x: x[1], reverse=False)
)

# Create the horizontal bar plot with two bars per crime (2022 and 2023)
plt.figure(figsize=(10, 6))

# Set the positions of the bars
bar_width = 0.35
index = np.arange(len(sorted_crimes))

# Plot the bars for 2022 and 2023 offenses using Seaborn's color palette
plt.barh(index - bar_width / 2, sorted_offenses_2022, bar_width, color=current_palette[8], label='2022')
plt.barh(index + bar_width / 2, sorted_offenses_2023, bar_width, color=current_palette[9], label='2023')

# Add annotations to each bar for clarity
for i, (value_2022, value_2023) in enumerate(zip(sorted_offenses_2022, sorted_offenses_2023)):
    plt.text(value_2022 + 10, i - bar_width / 2, f'{value_2022:.1f}', va='center', color='black', fontweight='bold')
    plt.text(value_2023 + 10, i + bar_width / 2, f'{value_2023:.1f}', va='center', color='black', fontweight='bold')

# Customize the plot
plt.xlabel('Offenses per 100,000 People', fontsize=12, fontweight='bold')
plt.ylabel('Type of Crime', fontsize=12, fontweight='bold')
plt.title('Crime Rates in 2022 and 2023 (Offenses per 100,000 People)', fontsize=14, fontweight='bold')
plt.yticks(index, sorted_crimes)
plt.legend()

# Grid lines for better readability
plt.grid(True, axis='x', linestyle='--', linewidth=0.7)

# Set x-axis range from 0 to 1600
plt.xlim(0, 1600)  # Set the x-axis range explicitly

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
