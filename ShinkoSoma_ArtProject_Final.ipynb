### Message


 **"Be the hope that lights the way."**

Women's education has come a long way, and today many women have access to primary education. However, the number of women who can pursue higher education remains a minority. The fewer the women who are able to continue their studies, the more challenging it becomes. Through this artwork, I want to send a message of encouragement to those women who continue to pursue education despite these challenges.

In this piece, I represent the percentage of women who complete secondary education with a gradient of blue—darker shades signify fewer women. The university enrollment rates for women are depicted as stars.

Just as the night sky appears darker, the stars shine brighter. The more visible you become, the more likely you are to face criticism or harassment. Nevertheless, stars continue to shine with strength and beauty in the night sky.

No one can take away your light. Be strong, be beautiful, and, while never forgetting the importance of those around you, trust in yourself and keep moving forward. You are the shining stars of hope for all.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Prepare numeric display (show even if NaN)
def display_value(value):
    return f"{value:.1f}" if not np.isnan(value) else "N/A"

# Create the plot (larger size)
fig, ax = plt.subplots(figsize=(18, 10))

# Draw each cell
for i in range(rows):
    for j in range(cols):
        country = world_grid[i][j]
        if country:
            value_uppersecondary = df_upper.loc[df_upper['alpha_3'] == country, 'latest_data_uppersecondary_female'].values
            value_bachelor_female = df_worldbank.loc[df_worldbank['alpha_3'] == country, 'latest_data_bachelor_female'].values

            display_text_uppersecondary = display_value(value_uppersecondary[0]) if len(value_uppersecondary) > 0 else "N/A"
            display_text_bachelor_female = display_value(value_bachelor_female[0]) if len(value_bachelor_female) > 0 else "N/A"

            # Draw a rectangle (cell for a country)
            rect = plt.Rectangle(
                (j, rows - i - 1), 1, 1, fill=True,
                facecolor="lightgray", edgecolor="gray"
            )
            ax.add_patch(rect)

            # Display country code and data (show country name even if data is missing)
            ax.text(j + 0.5, rows - i - 0.5,
                    f"{country}\n{display_text_uppersecondary}\n{display_text_bachelor_female}",
                    ha="center", va="center", fontsize=6,
                    fontweight="bold", color="black")

# Hide axes
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(0, cols)
ax.set_ylim(0, rows)
ax.set_frame_on(False)

# Background color
fig.patch.set_facecolor('#5A4FCF')
fig.patch.set_alpha(0.5)

# Title
plt.title("Upper Secondary Graduation Rate and Bachelor Enrollment (Female)", fontsize=20,
          color="white", fontweight="bold", pad=30)

# Legend text
ax.text(1, rows + 0.5,
        "1st line: Country (Alpha-3 Code)\n2nd line: Upper Secondary Graduation Rate (female)\n3rd Line: Bachelor Enrollment Rate (female)",
        ha="left", va="top", fontsize=12, color="white", fontweight="bold")

# Message
ax.text(cols - 7, 6, '"Be the hope \nthat lights the way."',
        ha="center", va="center", fontsize=18, color="white", fontweight="bold", fontstyle="italic")

# Show plot
plt.show()
```
