```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for average retail prices over time
years = np.arange(2008, 2015)
product_prices = {
    "A": [380, 370, 350, 320, 300, 280, 260],
    "B": [360, 350, 340, 310, 290, 270, 250],
    "C": [200, 220, 250, 270, 280, 290, 295],
    "D": [180, 200, 220, 240, 250, 260, 270],
    "E": [150, 180, 200, 210, 220, 230, 240]
}

# Colors for differentiation
colors = {"A": "blue", "B": "red", "C": "green", "D": "orange", "E": "purple"}

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each product's price trend
for product, prices in product_prices.items():
    ax.plot(years, prices, marker="o", linestyle="-", label=f"Product {product}", color=colors[product])

# Highlighting Product C launch in 2010
ax.axvline(x=2010, color='gray', linestyle="--", alpha=0.5)
ax.text(2010.2, 310, "Product C Launch", fontsize=12, color='black')

# Formatting
ax.set_title("Average Retail Price Over Time", fontsize=14)
ax.set_xlabel("Year")
ax.set_ylabel("Price ($)")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.5)

# Show the plot
plt.show()
1 + 1
```
