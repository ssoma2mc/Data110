### Task 1-1: Brief description of the Titanic dataset

The Titanic dataset contains information about passengers aboard the ill-fated RMS Titanic. It includes both personal details and survival information for each passenger. Key columns in the dataset include:

- **PassengerId**: A unique identifier for each passenger.
- **Pclass**: The class of the passenger (1, 2, or 3), representing the socio-economic status (1st, 2nd, or 3rd class).
- **Name**: The full name of the passenger.
- **Sex**: The gender of the passenger (male or female).
- **Age**: The age of the passenger in years.
- **SibSp**: The number of siblings or spouses the passenger had aboard the Titanic.
- **Parch**: The number of parents or children the passenger had aboard.
- **Ticket**: The ticket number.
- **Fare**: The amount of money the passenger paid for the ticket.
- **Cabin**: The cabin number where the passenger stayed (if available).
- **Embarked**: The port where the passenger boarded the Titanic (C = Cherbourg, Q = Queenstown, S = Southampton).
- **Survived**: Whether the passenger survived (1) or did not survive (0).

The dataset we use is from GitHub, which does not contain the data dictionary. However, the Titanic dataset is well-known, as it was used in the Kaggle competition. I used the dictionary from the Kaggle site.

---

### Data dictionary

| **Variable** | **Definition**                                   | **Key**                                               |
|--------------|--------------------------------------------------|-------------------------------------------------------|
| survival     | Survival                                         | 0 = No, 1 = Yes                                       |
| pclass       | Ticket class                                     | 1 = 1st, 2 = 2nd, 3 = 3rd                            |
| sex          | Sex                                              |                                                       |
| age          | Age in years                                     |                                                       |
| sibsp        | # of siblings / spouses aboard the Titanic       |                                                       |
| parch        | # of parents / children aboard the Titanic       |                                                       |
| ticket       | Ticket number                                    |                                                       |
| fare         | Passenger fare                                   |                                                       |
| cabin        | Cabin number                                     |                                                       |
| embarked     | Port of Embarkation                              | C = Cherbourg, Q = Queenstown, S = Southampton        |

---

### Variable Notes:
- **pclass**: A proxy for socio-economic status (SES):
  - 1st = Upper
  - 2nd = Middle
  - 3rd = Lower
- **age**: Age is fractional if less than 1. If the age is estimated, it is in the form of xx.5.
- **sibsp**: The dataset defines family relations in this way...
  - Sibling = brother, sister, stepbrother, stepsister
  - Spouse = husband, wife (mistresses and fiancés were ignored)
- **parch**: The dataset defines family relations in this way...
  - Parent = mother, father
  - Child = daughter, son, stepdaughter, stepson
  - Some children traveled only with a nanny, therefore parch=0 for them.

### Additional Insights from the Bar Graph, Line Plot, and Scatter Plot

The bar graph, line plot, and scatter plot offer several advantages over histograms and KDE plots, providing more detailed insights into the relationships between different variables:

1. **Bar Graph**:  
   The bar graph allows for a clear comparison of categorical data, such as survival rates across passenger classes. Unlike histograms that represent distributions, the bar graph provides a straightforward comparison that can show how survival rates vary across discrete groups (e.g., 1st-class vs. 3rd-class passengers).

2. **Line Plot**:  
   The line plot reveals trends over a continuous variable, such as age. It helps to identify patterns such as how survival rates change with age, showing that younger passengers, particularly children, had higher survival rates. This trend might not be as easily visible in a histogram or KDE, where the data might be more dispersed or smoothed out.

3. **Scatter Plot**:  
   The scatter plot helps visualize the relationship between two continuous variables, such as age and fare. It enables us to assess if there’s any correlation between these variables and survival. For instance, the scatter plot might show if passengers who paid higher fares had a higher chance of survival, providing a direct visual correlation that could be missed in histograms or KDE plots.

In summary, while histograms and KDE plots give a good understanding of the distributions, bar graphs, line plots, and scatter plots help highlight specific relationships, trends, and comparisons between variables, making them useful for drawing more nuanced conclusions from the dataset.
